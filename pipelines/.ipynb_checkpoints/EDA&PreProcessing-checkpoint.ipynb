## Importing Training Dataset

**Observation:** I had to manually add two leading zeros on line 577, how could I make sure all numbers have leading zeros if necessary?

```python
import pandas as pd

# importing training dataset
X_train = pd.read_csv("../dataset files/training/madelon_train.data", 
                      header=None, index_col=None, delimiter=' ')

# importing dataset targets
y_train = pd.read_fwf("../dataset files/training/madelon_train.labels", header=None, index_col=None)
 
X_train.info()
print('\n')
X_train.head()
```

```python
# describe for training dataset
X_train.describe()
```

## Normalizing dataset

Since this is a sintetic dataset, there's no harm in normalizing the data for every single pipeline. (no real loss of information over non-normalized data.)

```python
# normalizing training dataframe
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = pd.DataFrame(
    min_max_scaler.fit_transform(X_train), 
        index=X_train.index, 
        columns=X_train.columns
    )

# adding target column
X_train_scaled['target'] = y_train
X_train_scaled.head()
```

## Exporting dataset

```python
X_train_scaled.to_csv(path_or_buf="../dataset files/training/post_eda_train.csv")
```
